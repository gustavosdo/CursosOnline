### 1. Séries estacionárias

São ditas estacionárias as séries que a média e a variância são constantes. De forma geral, portanto, séries que apresentem componentes de tendência ou ciclo **não** são estacionárias. Assim sendo, séries estacionárias geralmente apresentam apenas componentes sazonais e/ou aleatórias.

Algumas técnicas analíticas de séries temporais são aplicáveis apenas em casos especiais de estacionariedade. Para aplicar, portanto, estes métodos a séries não-estacionárias precisaremos transformar estas séries.

Para determinar a estacionariedade de uma série utilizamos alguns testes:

#### <a href="https://en.wikipedia.org/wiki/Dickey%E2%80%93Fuller_test">Teste de Dickey-Fuller</a>

#### <a href="https://en.wikipedia.org/wiki/KPSS_test">Teste KPSS</a>

#### <a href="https://en.wikipedia.org/wiki/Phillips%E2%80%93Perron_test">Teste de Philips-Perron</a>

### 2. Médias móveis

É um processo de transformação utilizado para suavizar a série removendo outliers/sazonalidade e, assim, promove a identificação da tendência. Há dois tipos principais de médias móveis: simples e exponencial. A **ordem** de uma média móvel é a quantidade de dados a serem considerados ao redor do dado atual que vão contribuir para o cálculo da média.

```python
# Implementando médias móveis -------------------------------------------------------------------------------------

# Importando biblioteca para leitura de dados em csv
import pandas as pd

# Importando biblioteca otimizada para cálculos vetoriais
import numpy as np

# Importando biblioteca de gráficos
from matplotlib import pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
# Importando dados
passengers = pd.read_csv("../Material/Dados/AirPassengers.csv")

# Regularizando dados
ts = passengers.set_index("Month")
```

```python
# Definindo a ordem ideal para a média móvel:
# A ordem ideal é aquela que, uma vez aplicada, leva ao menor valor de erro entre o valor original
# e o valor da média móvel. Aqui utilizamos o erro quadrático médio (MSE) como métrica.

opt = None # ordem ideal
min_mse = None # erro absoluto médio no caso em que n = ordem ideal

# Convertendo valores dos dados para precisão float
db = ts[['#Passengers']].values.astype('float32')

# Resultados de MSE para todos os valores possíveis de n, ou seja, de 3 a len(db)/2 + 1
res = np.zeros(int(len(db)/2 - 2))

# Para n > len(db)/2 + 1 nenhum valor será diferente de NA
for n in range(3, int(len(db)/2 + 1)):
    # valores de MSE por ponto da série temporal
    iter_res = np.zeros(len(db) - n)
    for i in range(0, len(db) - n):
        iter_res[i] = np.power(np.mean(db[:,0][i:i+n]) - db[i+n][0], 2)
    # média dos MSE da série temporal para o dado n
    res[n-3] = np.mean(iter_res)

# encontrando a ordem ideal
opt = np.argmin(res) + 3
# e o menor mse (associado a ordem ideal)
min_mse = np.min(res)

print("MSE = %s" %res)
print("Melhor MSE = %s" %min_mse)
print("Ótimo n =%s" %opt)
```

```python
# Avaliação visual da ordem ideal
forecast = np.zeros(len(db) + 1) # dados com média móvel
# loop sobre os valores iniciais até o n ideal
for i in range(0, opt):
    forecast[i] = db[i][0] # mantemos os dados originais
# loop sobre os dados a partir do n ideal
for i in range(0, len(db) - opt + 1):
    # preenchemos de 0+3 até len(db) com a média móvel
    forecast[i+opt] = np.mean(db[:,0][i:i+opt])

plt.plot(db[:,0], label = "Dados Originais")
plt.plot(forecast, label = "Média móvel")
plt.legend()
plt.show()
```

```python
# Extra: Usando médias móveis para remover a sazonalidade
opt = 12
forecast = np.zeros(len(db) + 1) # dados com média móvel
# loop sobre os valores iniciais até o n ideal
for i in range(0, opt):
    forecast[i] = db[i][0] # mantemos os dados originais
# loop sobre os dados a partir do n ideal
for i in range(0, len(db) - opt + 1):
    # preenchemos de 0+3 até len(db) com a média móvel
    forecast[i+opt] = np.mean(db[:,0][i:i+opt])

plt.plot(db[:,0], label = "Dados Originais")
plt.plot(forecast, label = "Média móvel")
plt.legend()
plt.show()
```

### 3. Componentes de uma série temporal

```python
# Gráfico da série com todas as suas componentes ------------------------------------------------------------------

# Importando dataset (com a corretaleitura da data)
dataparse = lambda dates_vector: pd.datetime.strptime(dates_vector, "%Y-%m")
ts = pd.read_csv("../Material/Dados/AirPassengers.csv",
                 parse_dates = ["Month"], index_col = "Month", date_parser = dataparse)

# Desenhando gráfico com título e nomes nos eixos
plt.plot(ts)
plt.title("Passsageirtos internacionais transportados por trimestre", fontsize = 24)
plt.ylabel("Passageiros")
plt.xlabel("Data")
plt.show()
```

```python
# Decomposição sazonal aditiva
dec = seasonal_decompose(ts, model = 'aditiva')
# Gráficos da decomposição
dec.plot()
plt.title("Decomposição aditiva")
plt.show()
```

Note que, apesar da semelhança entre os dois resultados de decomposição, as escalas são bem distintas nos gráficos de resíduos e de sazonalidade. Neste caso em particular o resultado da decomposição multiplicativa é mais adequando já que podemos ver que a sazonalidade é mais pronunciada quanto mais passa o tempo.

### 4. Teste de estacionariedade

```python
# Importando teste de estacionariedade
from statsmodels.tsa.stattools import adfuller

# Separando valores da série temporal
X = ts['#Passengers']

# Aplicando teste de Dickey-Fuller
res = adfuller(X)

# Imprimindo resultados
print("Estatística ADF: %f" %res[0])
print("Valor de p: %f" %res[1])
print("Valores críticos:")
for key, value in res[4].items():
    print("\t%s: %.3f" % (key, value))
```

#### Interpretação:
A hipótese original do teste ($H_0$) aumentado de Dickey-Fuller é que a série temporal pode ser representada por uma função dependente do tempo (portanto não-estacionária). A hipótese alternativa, que rejeita a hipótese original, é que a série temporal é estacionária.

Nós interpretamos o resultado do teste (principalmente) por meio do valor-p. Dado que o valor-p do teste não ultrapasse um patamar (como 5%) sugere que nós rejeitemos a hipótese original em favor da hipótese alternativa.

O valor de Estatística ADF indica a probabilidade da hipótese original estar correta. Quanto maior, mais chance de termos a hipótese original verdadeira.

Os valores críticos indicam as probabilidades de considerar a dependência do tempo como uma flutuação estatística.

### 5. Transformação de séries

Agora iremos discutir como transformar uma série não-estacionária em estacionária.

```python
# Usar logaritmo pode ser útil
ts_log = np.log(ts)
ts_log.plot()
ts.plot()
plt.show()
```

Neste caso, mesmo com a visível suavização, não foi possível transformar a série em estacionária. Podemos realizar o teste aumentado de Dickey-Fuller para comprovar isso.

```python
# Separando valores da série temporal
X = ts_log['#Passengers']

# Aplicando teste de Dickey-Fuller
res = adfuller(X)

# Imprimindo resultados
print("Estatística ADF: %f" %res[0])
print("Valor de p: %f" %res[1])
print("Valores críticos:")
for key, value in res[4].items():
    print("\t%s: %.3f" % (key, value))
```

Ou seja, valor-p é muito maior que 5% (portanto temos dependência temporal).

Passamos então a diferenciação.

```python
# Tomando os valores da série original
y = ts["#Passengers"]

# Teste de Dickey-Fuller aumentado
def adf_test(y):
    print("Resultado do teste de Dickey-Fuller:")
    dftest = adfuller(y, autolag = "AIC")
    dfout = pd.Series(dftest[0:4], index = ["Teste", "Valor-p", "# de lags", "# de observações"])
    for key, value in dftest[4].items():
        dfout["Valor crítico ({})".format(key)] = value
    print(dfout)

# Testando a série original
adf_test(y)
```

```python
# Diferenciação de primeira ordem
dydt = np.diff(y)

# Teste da série de DF
adf_test(dydt)
```

Verifica-se que agora temos um valor-p muito próximo a 5% ainda que seja maior. Por outro lado, como parte dos resultados, vemos que a Estatística ADF tem valor menor do que o valor crítico de 5%. Isso sugere que podemos rejeitar a hipótese original com um nível de significância menor do que 5%, isto é, baixa probabilidade de que o resultado se deve a flutuação estatística.

Para termos um resultado mais ideal podemos diferenciar mais uma vez a série temporal.

```python
# Diferenciação de segunda ordem
dydt2 = np.diff(dydt)

# Teste de DF
adf_test(dydt2)
```

```python
# Gráficos da série temporal e suas derivadas
plt.plot(ts['#Passengers'])
plt.show()
```

```python
plt.plot(dydt, label = '1ª ordem')
plt.plot(dydt2, label = '2ª ordem')
plt.legend()
plt.show()
```

#### Diferenciação com Logaritmo e Média Móvel

```python
# Média e desvio-padrão móvel
rmean = ts.rolling(window = 12).mean()
rstd = ts.rolling(window = 12).std()
print(rmean)
print(rstd)
```

```python
# Aplicando média-móvel à amostra com logaritmo
ts_log_rmean = ts_log.rolling(window = 12).mean()
ts_log_rstd = ts_log.rolling(window = 12).std()

# Gráficos da amostra com log e log+media-movel
#plt.plot(ts, label = 'original')
plt.plot(ts_log, label = 'log')
plt.plot(ts_log_rmean, label = 'log+rmean')
plt.show()
```

Ainda observamos uma tendência. Na verdade, observamos a tendência nos dois gráficos mas apenas o gráfico com log apresenta tanto a média quanto a sazonalidade. Portanto se subtrairmos os dados com log e média-móvel dos dados com apenas log, esperamos observar um gráfico apenas sazonal mas sem tendência.

```python
# Subtraindo dados
ts_sub = ts_log - ts_log_rmean

# Removendo NA´s
ts_sub.dropna(inplace = True)
ts_sub.head()

# Gráfico da série
plt.plot(ts_sub)
```

```python
# Criando nossa própria função de teste de estacionariedade (com gráficos)
def test_stat(ts):
    # Calculando média-móvel e desvio-padrão-móvel
    rmean = ts.rolling(window = 12).mean()
    rstd = ts.rolling(window = 12).std()
    
    # Gráfico da série temporal, média e desvio-padrão
    ts_plot = plt.plot(ts, color = 'blue', label = 'Original')
    ts_rm_plt = plt.plot(rmean, color = 'red', label = 'Média-móvel')
    ts_std_plt = plt.plot(rstd, color = 'black', label = 'Desvio-padrão')
    plt.legend(loc = 'best')
    plt.title('Média-móvel e desvio-padrão')
    
    # Teste de estacionariedade
    adf_test(ts['#Passengers'])
    
# Função atua sobre os dados de passageiros
test_stat(ts_sub)
```

#### Remoção de tendência com regressão

```python
# Preparando dados para entrada na regressão linear
X = [i for i in range(len(ts))]

# Todos os valores de X tem que estar dentro de um vetor com apenas uma entrada
X = np.reshape(X, (len(X), 1))

# Número de passageiros
Y = ts.values
```

```python
# Importando bilioteca de regressão linear
from sklearn.linear_model import LinearRegression as lm

# Criando modelo de regressão linear
mod = lm(); mod.fit(X,Y)

# Criando reta da regressão
trend = mod.predict(X)

# Gráficos da série temporal e sua tendência (dada por regressão linear)
plt.plot(Y, label = 'Série temporal')
plt.plot(trend, label = 'Tendência')
plt.legend(loc = 'best')
plt.show()
```

```python
# Subtraindo a tendência da série temporal
detrended_ts = Y - trend

# Gráfico da remoção de tendência
plt.plot(detrended_ts)
```
